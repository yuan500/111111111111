````python
# Step 4: 下载APK
import glob, os
from google.colab import files
for apk in glob.glob('**/*.apk', recursive=True):
    print(f'{apk} ({os.path.getsize(apk)/1024/1024:.1f}MB)')
    files.download(apk)
````
